```python
corpus = """I am passionate about Artificial Intelligence and love learning new AI technologies. I enjoy building smart solutions and exploring Machine Learning, NLP, and Deep Learning."""
```

#### **_Tokenization_** - **_Paragraphs into Sentences_**

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize
```

```python
print(corpus)
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

#### **_Tokenization_** - **_Paragraphs or Sentence into Words_**

```python
from nltk.tokenize import word_tokenize
```

```python
paraWord = word_tokenize(corpus)
```

```python
paraWord
```

#### **_Sentence to Words_**

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

#### **_Sentence to Words_** - **_Split based on punctuations_**

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

#### **_Sentence to Words_**

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```

### **_Lemmatization_**

**_Lemmatization technique is like stemming. The output we will get after lemmatization is called ‘lemma’, which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing._**

#### **_Wordnet Lemmatizer_**

**_NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This class uses morph() function to the WordNet CorpusReader class to find a lemma._**

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going")
```

**_POS_**

* **_Noun - n_**
* **_Verb - v_**
* **_Adjective - a_**
* **_Adverb - r_**

```python
lemmatizer.lemmatize("going", pos='n')
```

```python
lemmatizer.lemmatize("going", pos='v')
```

```python
lemmatizer.lemmatize("going", pos='a')
```

```python
lemmatizer.lemmatize("going", pos='r')
```

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalized",]
```

```python
for word in words:
    print(word + "---" + lemmatizer.lemmatize(word, pos='v'))
```

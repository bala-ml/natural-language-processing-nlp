#### **_Stemming_**

**_Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP)._**

* **_Classification Problem_**
* **_Comments of product is a positive review or negative review_**
* **_Reviews - eating, eat, eaten - eat, going, gone, goes - go_**

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalized",]
```

#### **_Porter Stemmer_**

```python
from nltk.stem import PorterStemmer
```

```python
porterStem = PorterStemmer()
```

```python
for word in words:
    print(word + "---" + porterStem.stem(word))
```

```python
porterStem.stem('congratulations')
```

```python
porterStem.stem('sitting')
```

#### **_Regexp Stemmer_**

**_NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example._**

```python
from nltk.stem import RegexpStemmer
```

```python
regxStem = RegexpStemmer()
```

```python
regxStem = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
regxStem.stem('eating')
```

#### **_Snowball Stemmer_**

**_Snowball Stemmer is an advanced version of the Porter Stemmer in NLTK. It supports multiple languages and provides more accurate stemming results compared to Porter Stemmer.It reduces words to their root/stem form by removing prefixes and suffixes_**

```python
from nltk.stem import SnowballStemmer
```

```python
snowballStem = SnowballStemmer()
```

```python
snowballStem = SnowballStemmer('english')
```

```python
for word in words:
    print(word + "---" + snowballStem.stem(word))
```

```python
porterStem.stem("fairly"), porterStem.stem("sportingly")
```

```python
snowballStem.stem("fairly"), snowballStem.stem("sportingly")
```

```python
snowballStem.stem("goes")
```
